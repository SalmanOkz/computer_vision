```python
import cv2
import matplotlib.pyplot as plt

# 1. Load two binary images
img1_path = "security_image1.png"  # Replace with your file name
img2_path = "security_image2.png"

img1 = cv2.imread(img1_path, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(img2_path, cv2.IMREAD_GRAYSCALE)

# Check that both images loaded correctly
if img1 is None or img2 is None:
    raise FileNotFoundError("One or both input images could not be loaded.")

# 2. Ensure the images are binary (0 or 255)
_, img1_bin = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY)
_, img2_bin = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)

# 3. Perform bitwise operations
and_result = cv2.bitwise_and(img1_bin, img2_bin)
or_result  = cv2.bitwise_or(img1_bin, img2_bin)

# 4. Save the processed images
cv2.imwrite("security_AND_result.png", and_result)
cv2.imwrite("security_OR_result.png", or_result)

print(" Bitwise AND and OR results saved successfully!")

# 5. (Optional) Display all images using matplotlib
plt.figure(figsize=(10,6))
plt.subplot(2,2,1); plt.imshow(img1_bin, cmap='gray'); plt.title("Image 1")
plt.subplot(2,2,2); plt.imshow(img2_bin, cmap='gray'); plt.title("Image 2")
plt.subplot(2,2,3); plt.imshow(and_result, cmap='gray'); plt.title("Bitwise AND (Common Areas)")
plt.subplot(2,2,4); plt.imshow(or_result, cmap='gray'); plt.title("Bitwise OR (Combined Areas)")
plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# 1. Read the two product images
img1_path = "product1.jpg"   # Replace with actual file path
img2_path = "product2.jpg"

img1 = cv2.imread(img1_path)
img2 = cv2.imread(img2_path)

# Check if both images loaded correctly
if img1 is None or img2 is None:
    raise FileNotFoundError("One or both product images could not be loaded.")

# 2. Resize the images to the same size
# We'll use the size of the first image as reference
height, width = img1.shape[:2]
img2_resized = cv2.resize(img2, (width, height))

# 3. Blend the images
# Alpha controls transparency: 0.5 means equal blending
alpha = 0.5
beta = 1 - alpha

blended = cv2.addWeighted(img1, alpha, img2_resized, beta, 0)

# 4. Save the blended image
cv2.imwrite("product_composite.jpg", blended)
print(" Blended image saved as 'product_composite.jpg'")

# 5. Display all images
plt.figure(figsize=(12,6))
plt.subplot(1,3,1); plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)); plt.title("Product Image 1")
plt.subplot(1,3,2); plt.imshow(cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB)); plt.title("Product Image 2 (Resized)")
plt.subplot(1,3,3); plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB)); plt.title("Blended Composite")
plt.tight_layout()
plt.show()
```

```python
import cv2

# 1. Load the image
image_path = "/content/images.jpeg"
image = cv2.imread(image_path)

if image is None:
    raise FileNotFoundError(" Image not found. Please check the file path.")

# 2. Define a Region of Interest (ROI)
# Coordinates format: [y1:y2, x1:x2]
# Example: Crop a rectangular area around a face manually
# (These coordinates should be adjusted based on the image)
x, y, w, h = 100, 80, 200, 200  # left, top, width, height
roi = image[y:y+h, x:x+w]

# 3. Save the cropped region
cv2.imwrite("cropped_face.jpg", roi)
print(" Cropped image saved as 'cropped_face.jpg'")

# 4. (Optional) Display the images
cv2.imshow("Original Image", image)
cv2.imshow("Cropped ROI", roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
