    *   `0`: This is `dy`, the order of the derivative in the y-direction. A value of `0` means no derivative along the y-axis for this calculation.

    *   `sobel_x`: The result, which represents the horizontal edges detected by the Sobel filter, is stored in this variable.

*   `sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)`:
    *   This line is similar to the `sobel_x` calculation, but here:
    *   `0`: `dx` is `0`, meaning no derivative along the x-axis.
    *   `1`: `dy` is `1`, meaning the first derivative along the y-axis is calculated.
    *   `sobel_y`: The result, representing the vertical edges, is stored in this variable.

    *   `plt.title("Original")`: This sets the title for this specific subplot to "Original".

*   `plt.subplot(1,3,2); plt.imshow(sobel_x, cmap='gray');plt.title("Sobel X")`:
    *   This creates the 2nd subplot (1 row, 3 columns, 2nd position).
    *   It displays the `sobel_x` image (horizontal edges) in grayscale.
    *   It sets the title for this subplot to "Sobel X".

*   `plt.subplot(1,3,3); plt.imshow(sobel_y, cmap='gray');plt.title("Sobel Y")`:
    *   This creates the 3rd subplot (1 row, 3 columns, 3rd position).
    *   It displays the `sobel_y` image (vertical edges) in grayscale.
    *   It sets the title for this subplot to "Sobel Y".

*   `plt.show()`:
    *   This command displays all the figures and plots that have been created. Without this, the plots might not appear.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/download.jpg', 0)

# Check if image was loaded successfully
if img is None:
    print("Error: Image not found or could not be loaded. Please check the path.")
else:
    sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

    plt.figure(figsize=(12,5))
    plt.subplot(1,3,1); plt.imshow(img, cmap='gray');plt.title("Original")
    plt.subplot(1,3,2); plt.imshow(sobel_x, cmap='gray');plt.title("Sobel X")
    plt.subplot(1,3,3); plt.imshow(sobel_y, cmap='gray');plt.title("Sobel Y")
    plt.show()
```

```python
magnitude = cv2.magnitude(sobel_x, sobel_y)

# Gradient Orientation
orientation = cv2.phase(sobel_x, sobel_y, angleInDegrees=True)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1); plt.imshow(magnitude, cmap='gray'); plt.title("Gradient Magnitude")
plt.subplot(1,2,2); plt.imshow(orientation, cmap='hsv');plt.title("Gradient Orientation")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure
import numpy as np
from PIL import Image

# Path to your image file (not a zip file)
image_path = '/content/images.jpg'

# Load the image directly in grayscale
img_leaf = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if image was loaded successfully
if img_leaf is None:
    print(f"Error: Image not found or could not be loaded from {image_path}. Please check the path.")
else:
    # Compute HOG features
    hog_features, hog_image = hog(img_leaf,
                                  orientations=9,
                                  pixels_per_cell=(8,8),
                                  cells_per_block=(2,2),
                                  visualize=True,
                                  block_norm='L2-Hys')

    # Improve visibility of HOG output
    hog_image_rescaled = exposure.rescale_intensity(hog_image,
                                                     in_range=(hog_image.min(), hog_image.max()))

    # Display original and HOG visualization
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    plt.imshow(img_leaf, cmap='gray')
    plt.title("Original Leaf Image")
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(hog_image_rescaled, cmap='gray')
    plt.title("HOG Features")
    plt.axis('off')
    plt.show()
```
